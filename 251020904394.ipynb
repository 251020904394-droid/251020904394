```python
n=int(input("Nhap so nguyen duong n: ",))
o=0
for i in range (n+1):
  if i%2==0:
    o+=i
print(o)
```

```python
n = input("Nhập chuỗi: ")
tungtu = n.split()
maxtu = ""
for tu in tungtu:
    if len(tu) > len(maxtu):
        maxtu = tu
print(len(maxtu), "tu dai nhat:", maxtu)
```

```python

songuyen = list(map(int, input("Nhap cac so nguyen: ").split()))
songuyen = list(set(songuyen))
songuyen.sort()
print("Danh sách sau khi xử lý:", songuyen)
if len(songuyen) < 2:
    print("Không tồn tại số lớn thứ hai")
else:
    print("Số lớn thứ hai là:", songuyen[-2])
```

```python
n=input(" ",)
nn=n.split()
tungdiem=list(map(int,nn))
diemtong=0
quamon=0
for i in tungdiem:
  if i>=5:
    quamon+=1
    diemtong+=i
diemtrungbinh=diemtong/quamon
print("Diem trung binh: ",diemtrungbinh ,"Qua mon: ",quamon)
```
